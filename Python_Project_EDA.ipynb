```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Diwali Sales Data.csv", encoding='latin1')
```

```python
df.shape
```

```python
df.describe()
```

```python
df.head()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace = True)
```

```python
df.head()
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace = True)
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = "Gender", palette= 'Set1')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen, palette = 'Set1')

# From this graph we can conclude that most of the buyers are females with a greater purchasing power as compared to males
```

```python
ax = sns.countplot(data = df, x = "Age Group", hue = "Gender")

for bars in ax.containers:
    ax.bar_label(bars)

# From this graph we can say that most of the buyers are from age group of 26-35 with the majority of females
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age, palette = 'Set1')
```

```python
# Total number of orders from top 5 states

sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(5)

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x = 'State', y = 'Orders', data = sales_state, palette = 'Set1')
```

```python
# Total Amount/Sales from top 5 states

sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(5)

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x = 'State', y = 'Amount', data = sales_state, palette = 'Set1')
```

```python
# On the basis of marital status

ax = sns.countplot(data = df, x = "Marital_Status", hue = "Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data = sales_marital, x = 'Marital_Status', y = 'Amount', hue = 'Gender')

# States that most buyers are married women with a greater purchasing pwer compared to men
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = "Occupation", palette = "Set2")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_oc = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(data = sales_oc, x = 'Occupation', y = 'Amount', palette = 'Set1')

# We can observe that most buyers are working in IT, Healthcare and Aviation sector.
```

```python
sns.set(rc={'figure.figsize':(26,5)})
ax = sns.countplot(data = df, x = "Product_Category", palette = "Set2")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_prod = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.barplot(data = sales_prod, x = 'Product_Category', y = 'Amount', palette = 'Set1')

# Therefore, most sold products are from - Food, Clothing and Electronics category
```

```python
sales_prod = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

sns.barplot(data = sales_prod, x = 'Product_ID', y = 'Orders', palette = 'Set1')
```

```python
## Insights ##
# 1. MOST BUYERS ARE MARRIED WOMEN BELONGING FROM THE AGE GROUP OF 26-35
# 2. Residing in UP, Maharashtra and Karnataka
# 3. Working in IT, Healthcare and Aviation sector
# 4. Spending mostly on Food, Clothing and Electronics category
```
